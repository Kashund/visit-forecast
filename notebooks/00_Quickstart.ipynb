# Visit Forecast – Quickstart

This notebook uses the same forecasting engine as the Streamlit app (`visit_forecast`).

**Data columns required:** `Date`, `Visits`, `Department`

A sample dataset is included at `data/sample/synthetic_visits.csv`.

```python
from visit_forecast import forecast_visits
import pandas as pd

# Option A: use included sample CSV
df = pd.read_csv('../data/sample/synthetic_visits.csv')

result = forecast_visits(
    df=df,
    timeframe_start='2020-04-01',
    timeframe_end='2025-03-01',
    forecast_periods=12,
    department='All',
    adjustment_percent=10,
    changepoint_prior_scale=0.05,
    interval_width=0.90,
)

result.performance_metrics
```
